```python
#  Problem Solving Assignment:
# 1 problem  

def reverse_words(s):
   
    words = s.split(" ")
    
    reversed_list = words[::-1]
    
    result = " ".join(reversed_list)

    return result

input_str = "The greatest victory"
print(reverse_words(input_str)) 
```

```python
#  Problem Solving Assignment:
# 2 problem 

def are_you_playing_banjo(name):
   
    if name[0].lower() == 'r':

        return name + " plays banjo"
    
    else:
        return name + " does not play banjo"

print(are_you_playing_banjo("Ramy"))   
print(are_you_playing_banjo("roland")) 
print(are_you_playing_banjo("Ahmed"))  
```

```python
#  Problem Solving Assignment:
# 3 problem 

def twoSum(nums, target):
    prevMap = {}
    for i, n in enumerate(nums):
        diff = target - n
        if diff in prevMap:
            return [prevMap[diff], i]
        prevMap[n] = i

input_list = input("Enter numbers separated by space: ")

nums = [int(x) for x in input_list.split()]


target = int(input("Enter the target: "))


result = twoSum(nums, target)
if result:
    print(f"The indices are: {result}")
else:
    print("No solution found.")
```
