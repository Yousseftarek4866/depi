```python
# Factorial numbers 
def factorial(num) : 
    if num == 0 : 
        return 1
    if num == 1 :
        return 1 
    return num * factorial(num -1 )
factorial(5)
```

```python
def try_factorial(num) :
    if num == 0 :
     return 1
    if num == 1 :
       return 1
    return num * try_factorial(num -1)
try_factorial(5)

    
 
```

```python
# prime numbers
def isprime(num) : 
  for i in range(2 , num) : 
    if num % i == 0: 
        return False 
  return True
isprime(7)
```

```python
def try_prime(num) : 
    for i in range(2 , num) : 
        if num % i == 0 : 
         return False 
    return True     
try_prime(19)
```

```python
 
def dividors(num1 , num2 ): 
    """
   takes 2 num : then gets the min between them 
   check if the 2 numbers have reminders or not 
   if they are divisible , values are append inside the empty list 

    """
    mn = min( num1 , num2 ) 
    common_dividors = [] 
    for i in range(1 ,mn+1) : 
       if num1 % i == 0 and num2 % i == 0 : 
           common_dividors.append(i)
    return common_dividors 
 
```

```python
def try_dividors (num1 , num2 ) : 
    mn = min(num1 , num2)
    common_numbers = []
    for i in range(1 , mn+1) : 
      if num1 % i == 0 and num2 % i == 0 :
       common_numbers.append(i) 
    return common_numbers 
try_dividors(12 , 8)


     
```
