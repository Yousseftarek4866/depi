# Task 1.

- Calculate Factorial using recursion


```python
class Factorial:
    """
    Factorial class handles the operation of calculating the factorial of a number.
  
    """

    def __init__(self, num):
        """
        Initializes the Factorial class with a number.
        """
        self.num = num

    def calculate(self):
        """
        The main method to trigger the factorial calculation.
        
      
        """
        return self._factorial(self.num)

    def _factorial(self, num):
        """
        A private-like method that uses recursion to find the factorial.
        
      
        """
        if num == 0 or num == 1:
            return 1
        return num * self._factorial(num - 1)


fact = Factorial(5)


print(f"Factorial of {fact.num} is: {fact.calculate()}")
```

# Task 2.

- Write a function that takes a number and return True if this number is prime and False if not.


```python
class PrimeChecker:
    """
    PrimeChecker class handles the operation of checking whether a number is prime.
    
    
    """

    def __init__(self, number):
        """
        Initializes the PrimeChecker class with a number.
        """
        self.number = number

    def is_prime(self):
        """
        Determines whether the number is prime or not.
        
        Returns:
            bool: True if the number is prime, otherwise False.
        """
        if self.number < 2:
            return False
        
        # Loop men 2 le 7ad el-raqam (exclusive)
        for i in range(2, self.number):
            if self.number % i == 0:
                return False
        return True


checker = PrimeChecker(20)


result = checker.is_prime()


print(f"Is {checker.number} a prime number? {result}")

checker2 = PrimeChecker(7)
print(f"Is {checker2.number} a prime number? {checker2.is_prime()}")
```

# Task 3.

- Write a function that takes two numbers and return a list of common dividors


```python

class CommonDivisors:
    """
    CommonDivisors class handles the operation of finding all common 
    divisors between two numbers.
    
    Attributes:
        num1 (int): The first number.
        num2 (int): The second number.
    """

    def __init__(self, num1, num2):
        """
        Initializes the CommonDivisors class with two numbers.
        """
        self.num1 = num1
        self.num2 = num2

    def find_divisors(self):
        """
        Finds all numbers that divide both num1 and num2 without a remainder.
        
        Returns:
            list: A list of common divisors.
        """
     
        mn = min(self.num1, self.num2)
        common_dividors = []
        
        for i in range(1, mn + 1):
            if self.num1 % i == 0 and self.num2 % i == 0:
                common_dividors.append(i)
                
        return common_dividors


task_obj = CommonDivisors(20, 10)

result = task_obj.find_divisors()


print(f"Common Divisors of {task_obj.num1} and {task_obj.num2} are: {result}")
```

```python
class EvenNumbers:
    """
    EvenNumbers class handles the operation of finding all even numbers 
    up to a certain limit.
    
    Attributes:
        limit (int): The upper bound to check for even numbers.
    """

    def __init__(self, limit):
        """
        Initializes the EvenNumbers class with a limit.
        """
        self.limit = limit
        self.even_list = [] 

    def get_even_numbers(self):
        """
        Iterates from 0 to the limit and finds even numbers.
        
        Returns:
            list: A list containing all even numbers found.
        """
  
        self.even_list = [] 
        
        for i in range(self.limit):
            if i % 2 == 0:
                self.even_list.append(i) 
        
        return self.even_list

task_obj = EvenNumbers(20)


result = task_obj.get_even_numbers()

print(f"Even numbers up to {task_obj.limit} are: {result}")
```

# Task 4.

- Given two strings, write a program that efficiently finds the longest common subsequence

- **Ex:** str1 = "Welcome to Machine Learning Diploma" str2 = "I am studying a Machine Learning Course" --> Machine Learning


```python
class LongestSubstring:
    """
    LongestSubstring class finds the longest common string between two given strings.
    
    """

    def __init__(self, str1, str2):
        """
        Initializes the class with two strings.
        """
        self.str1 = str1
        self.str2 = str2

    def find_longest(self):
        """
        Iterates through all possible substrings to find the longest common one.
        
        """
        longest_str = ""
       
        for i in range(len(self.str1)):
            for j in range(i, len(self.str1)):
                sub_str = self.str1[i : j + 1]

              
                if sub_str in self.str2:
                    
                    if len(sub_str) > len(longest_str):
                        longest_str = sub_str
                        
        return longest_str


checker = LongestSubstring("Amit Learning", "I am learning at Amit")


result = checker.find_longest()

print(f"The longest common substring is: '{result}'")
```

# Task 5.

- Write a function that will take a given string and reverse the order of words.

- **Ex:** “Hello world” → ”world Hello”


```python
class SentenceReverser:
    """
    SentenceReverser class handles the operation of reversing the order 
    of words in a given sentence.
    
    
    """

    def __init__(self, sentence):
        """
        Initializes the class with the target sentence.
        """
        self.sentence = sentence

    def reverse_words(self):
        """
        Splits the sentence into words, reverses their order, and joins them back.
        
        """
       
        words_list = self.sentence.split(" ")
        
 
        reversed_list = words_list[::-1]
        
        
        reversed_sentence = " ".join(reversed_list)
        
        return reversed_sentence

input_text = "Hello World"
reverser = SentenceReverser(input_text)

result = reverser.reverse_words()


print(f"Original: {input_text}")
print(f"Reversed: {result}")
```

# Task 6.

- Create a program that generates a random password of a given length.
- Hint: use random module → import random → random.randrange


```python
import random

class PasswordGenerator:
    """
    PasswordGenerator class handles the generation of a random secure password.
    
  
    """

    def __init__(self, length):
        """
        Initializes the class with the required password length.
        """
        self.length = length

    def generate_password(self):
        """
        Generates a random password using letters, numbers, and symbols.
        
     
        """
        characters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*"
        num_of_chars = len(characters)
        password = ""

       
        for _ in range(self.length):
            
            rand_indx = random.randrange(0, num_of_chars)
            password += characters[rand_indx]

        return password


if __name__ == "__main__":

    p1 = PasswordGenerator(12)

    generated_pass = p1.generate_password()
    print(f"Your new password is: {generated_pass}")
```

- Write a Python program to find all the unique words and count the frequency of occurrence from a given list of strings

- **Ex:** ["Welcome", "Ali", "Hi", "Ali", "No", "Hi", "No", "Ali", "No", "Ali"] → {'Ali': 4, 'Welcome': 1, 'No': 3, 'Hi': 2}


```python
class WordCounter:
    """
    WordCounter class handles the operation of counting the frequency 
    of each word in a given list.
    
    
    """

    def __init__(self, words_list):
        """
        Initializes the WordCounter class with a list of words.
        """
        self.words_list = words_list

    def count_frequencies(self):
        """
        Calculates how many times each unique word appears in the list.
        
      
        """
  
        unique_words = set(self.words_list)
        
      
        counts = {word: 0 for word in unique_words}
        

        for i in unique_words:
            for j in self.words_list:
                if i == j:
                    counts[i] += 1
                    
        return counts


if __name__ == "__main__":
   
    words = ["Welcome", "Ali", "Hi", "Ali", "No", "Hi", "No", "Ali", "No", "Ali"]
    counter = WordCounter(words)

    result = counter.count_frequencies()

   
    print("Word Frequencies:")
    for word, count in result.items():
        print(f"{word}: {count}")
counts
```

```python
class WordCounter:
    """
    WordCounter class handles the operation of counting the frequency 
    of each word in a given list using built-in list methods.

    """

    def __init__(self, words_list):
        """
        Initializes the WordCounter class with a list of words.
        """
        self.words_list = words_list

    def get_frequencies(self):
        """
        Calculates the count of each unique word in the list.
        
        
        """
       
        unique_words = set(self.words_list)
        

        counts = {word: self.words_list.count(word) for word in unique_words}
        
        return counts


if __name__ == "__main__":
   
    my_words = ["Welcome", "Ali", "Hi", "Ali", "No", "Hi", "No", "Ali", "No", "Ali"]
    counter = WordCounter(my_words)

   
    result_dict = counter.get_frequencies()

  
    print("Word Frequency Analysis:")
    for word, count in result_dict.items():
        print(f"{word}: {count}")
```

# Task 9.

- Implement a basic chatbot that can respond to user input with predefined responses


CONVERT IT TO CLASS STRUCTURE


```python
import random

class SimpleChatbot:
    """
    SimpleChatbot class handles a basic interaction system using predefined responses.
    
    Attributes:
        responses (dict): A dictionary mapping keywords to lists of possible replies.
    """

    def __init__(self):
        """
        Initializes the chatbot with a set of predefined responses.
        """
        self.responses = {
            "hello": ["Hello!", "Hi there!", "Greetings!"],
            "how are you": ["I'm doing well, thank you!", "I'm fine, how about you?"],
            "goodbye": ["Goodbye!", "See you later!", "Farewell!"],
            "default": ["I'm sorry, I didn't understand.", "Could you please rephrase that?"]
        }

    def get_response(self, user_input):
        """
        Analyzes user input and returns a random matching response.
        
        Args:
            user_input (str): The text entered by the user.
            
        Returns:
            str: A random response from the matching keyword list.
        """
        user_input = user_input.lower()
        
        for key in self.responses:
            if key in user_input:
                return random.choice(self.responses[key])
        
        return random.choice(self.responses["default"])

    def start_chat(self):
        """
        Starts the chatbot loop to interact with the user via console.
        """
        print("Chatbot: Hi! How can I assist you today? (Type 'goodbye' to exit)")
        
        while True:
            user_input = input("User: ")
            response = self.get_response(user_input)
            print("Chatbot:", response)
            
            if "goodbye" in user_input.lower():
                break

if __name__ == "__main__":
  
    my_bot = SimpleChatbot()

    my_bot.start_chat()
```
