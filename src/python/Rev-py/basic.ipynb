```python
# the Average of grades 
def calculate_average():
    num_st=int(input('number of students'))
    grades = []

    for i in range(num_st) : 
        grade = float(input('enter the grade'))
        grades.append(grade)
    if num_st == 0 :
        print('no students')
    else : 
        print( f'the average of grades  = {sum(grades)/ num_st} ' ) 

calculate_average()
```
